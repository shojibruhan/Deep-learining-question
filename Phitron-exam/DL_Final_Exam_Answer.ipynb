## Explain the concept of backpropagation in detail. Describe how the error is propagated from the output layer to the hidden layers. Derive the mathematical expressions for weight updates using the chain rule, and explain how learning rate affects the training process.


## Write Answer 01:

**Main Concept behind backpropagation:**
---
___

Backpropagation is the heart of neural network. Core concept of it is to tuning the weight and bias to reduce the loss to minimum. Let's say the loss function L is:

$$ L= {(y - \widehat{y}){^2}} $$

During the forward pass, we assign random value for w and b. Which produce error. Backpropagation helps us to find suitable value for w and b so that L can be as minimum as possible. For this it use partial derivative with respect to w and b.


**Error Prpagation:**
---
___
Errors are progated from output layers to hidden layers through backpropagation. Chain rule is used for this pupose.

After forward pass we masaured the error by comparing true value with predicted value. We use some function call loss function like Mean Squared Error we used above.

With the help of gradient descent algorithm we calculate the gradient of the error with respect to output value from output layers.

This process repeats until it reach input layers when it figure out how much weight and bias need to be changed to reduce the loss.



**Mathematical Expressions:**
---
___

Weight update rule:
$$ w_{new}= w_{old} - η\frac{∂L}{∂w} $$

Loss function:
$$ L= (y - \widehat{y})^2 $$

Here,
$$ w_{new}= new\ weight $$
$$ w_{old}= old\ weight $$
$$ η= learning\ rate $$
$$ L= Loss $$
$$ x= input \ value $$



L changes with prediction ŷ

$$ \frac{∂L}{∂\widehat{y}}= \frac{∂}{∂\widehat{y}}(y - \widehat{y})^2 $$

$$ = -2(y - \widehat{y})$$

Again, ŷ dipends on 𝑤. So,

$$ \frac{∂\widehat{y}}{∂w}= \frac{∂}{∂w}(wx + b) $$
$$ = x \ (input) $$

So using chain rule,

$$ \frac{∂L}{∂w} = \frac{∂L}{∂\widehat{y}}  * \frac{∂\widehat{y}}{∂w}  $$
$$ = -2(y - \widehat{y}) * x $$

new weight,

$$ w_{new}= w_{old} - η * {\{ -2(y - \widehat{y}) * x }\} $$


**Affects of learning rate in training process:**
---
___
Learning rate controlls the update step in backpropagation. It can't be too big or too small. Should be moderate.

> if η get too bigger:

* Oscilate too much. Can't converge.
* causing exploding gradient problem

> if η get too smaller:

* Training going to very slow. Take lifetime to converge.
* Can stuck in local minima.
* cause vanishing gradient problem


> Ideal learning rate:
* Update is balaced.
* Convergence is faster

## Discuss the vanishing gradient problems in deep neural networks. Explain why they occur during backpropagation, how activation functions contribute to this issue, and suggest at least three techniques to overcome these problems.


## Write Answer 03:

**Vanishing Gradient Problems**
---
Instead of update accordingly, weights tend to update extremly small in backpropagation causing the model stop learning. This ks called vanishing gradient problems. Models fail to capture important features.

**Why vanishing gradient problems occur in backpropagation**
---
The weight update formula is:
$$ w_{new}= w_{old} - η\frac{∂L}{∂w} $$

Here,
$$ w_{new}= new\ weight $$
$$ w_{old}= old\ weight $$
$$ η= learning\ rate $$
$$ L= Loss $$

To update weights, we have to multiplied the gradient accross layers using chain rule.

$$ \frac{∂L}{∂w} = \frac{∂L}{∂z_{i}}  * \frac{∂z_{i}}{∂z_{i-1}} * \frac{∂z_{i-1}}{∂z_{i-2}} * \ . \ . \ .  $$

If the gradient become smaller and smaller, ultimate products also getting smaller. Which makes the gradient to shrink.

**Role of Activation function**
---

Activation function plays an important role. Specialy sigmoid or Tanh can cause vanishing gradient problems. Output range of sigmoid in 0 and 1. And for Tanh it's -1 to 1. So if the value is less then 1, a decimal/ fraction, gradiants tends to shrink as depth increases.


**3 Ways to overcome vanishing gradient problems**
---
* `Reduce the complexity`
  If the layers getting deep, sometimes it shows vainishg gradient problem. Best way is to reduce the number of layer.

* `RelU activation function` Output range of ReLU is
$$ f(x)= max(0, x) $$
ReLU prevents from shrinking. It's one of the best way to prevening vanishing gradient problem.

* `Batch Normalization` Normalizing each layer input can stabilize the gradient throughout the operation.

## Explain different activation functions used in neural networks, including Sigmoid, Tanh, and ReLU. Provide their mathematical formulas, advantages, disadvantages, and suitable use cases. Also discuss why ReLU is widely used in deep learning.

## Write Answer 06:

After taking an input, we added weight and bias to the input. Then we do weighted sum operation where we add the wigthed input with bias.

$$ z= w_ix_i + b $$

The z is a raw output. To add non linearity we use a mathematical function. It decide if the neuron should activate or not. If activated how much? This is called activation function. There are so many type of activation function like:

* Sigmoid
* Tanh
* ReLU
* Softmax  
and many more.


**Sigmoid**
---
___

Sigmoid is the primary activation function which introduce the non linearity to the output value. It is 'S' shaped curve produces valus between 0 and 1. It is widely use in binary classification problem.


<u>*_Formula_*</u>


$$ \sigma(x)=  \frac{1}{1+e^{-x}} $$


$$ \lim\limits_{x→\ +\ ∞} σ(x) = 1 $$


$$ \lim\limits_{x→\ -\ ∞} σ(x) = 0 $$


Here,
$$ x= input features $$
$$ σ(x)= output \ value $$
$$e= Euler's \ numbe(≈ 2.718)$$


<u>*_Advantage_*</u>

* Smooth gradients. Best for gradient based optimization methods.
* Output can interpret directly as probality.

* Can create clear prediction since the ouput value of sigmoid is 0 and 1.

* globally differentiable.


<u>*_Disadvantage_*</u>


* Can create vanishing gradient problem. When it happens models stop to learn or learn very slowly.

* Outputs are non zero centered. So optimization get harder.

* Learning dynamics get slower if input size is bigger.


<u>*_use cases_*</u>

* Binary classification like spam detection ('spam' or 'not spam'), medical diagnosis('healthy' or 'not healthy').

* Multi label classification problem where a single instance can belong to multiple class. Sigmoid can apply to each individual neuron to classify the object.

* logistic regression, where a linear score can convert into probality score.

**Tanh**
---
___


Hyperbolic Tangent function or Tanh is a non-linear 'S' shaped function. It produces ouput between -1 and 1. It is superior than sigmoid since it is zero centered means it keeps mean of activation distribution 0. It also keeps optimization stable.



*_Formula_*

$$ \tanh(x)=  \frac{e^{x}+e^{-x}}{e^{x}+e^{-x}} $$

$$ -1 < tanh(x) < 1 $$


$$ \lim\limits_{x→\ +\ ∞} tanh(x) = 1 $$


$$ \lim\limits_{x→\ -\ ∞} tanh(x) = - 1 $$

<u>*_Advantage_*</u>

* zero centerd activation.
* Compared to sigmoid, tanh has stronger gradients.
* smooth and differentiable everywhere.



<u>*_Disadvantage_*</u>
* Like sigmoid, here also can happen vanishing gradient problem. If the value is less then 1 gradient can shrink.

* Can't prevent gradient collapse in deep architecture.

* Since it use exponential function, computationally can expensive.

<u>*_use cases_*</u>

* tanh is important where negetive values are also significant.

* Widely use in Recurrent Neural Network specifically in Long-Short Term Memory (LSTM).

* In Generative Adversarial Network(GANs) tanh is used as final activation function.

**ReLU**
---
___

Rectified Linear Unit, in short ReLU is widely used standerd activation function. It ouputs directly the inputs if inputs are positive, otherwise provides 0. That is it surpress the whole activation for negetive half space.

It is better in efficient computing and handling gradient than sigmoid and tanh. ReLU solves the vanishing gradient problem during backpropagation. Also it doesn't use exponential function like previous two. So computational time also get reduced.




<u>*_Formula_*</u>

$$ ReLU(x)= \text{max}(0, x) $$

$$ ReLU(x) =  \begin{cases}0 & \text{if } x < 0 \\ x & \text{if } x \geq 0  \end{cases} $$


<u>*_Advantage_*</u>

* Solve vanishing gradient problem.

* Computationally inexpensive.

* Empirical convergence is faster in deep architecture.

<u>*_Disadvantage_*</u>

* It's not zero centered.

* Dead neuron phenomenon. Neuron effectively stops learning.

<u>*_use cases_*</u>

* ReLU is now Industry standerd activation function in deep neural network.

* Default activation function in Convolutional Neural Network to extract visual features.

* Use in Multi Layer Perceptron to solve vanishing gradient problem.

* Regression model whwere predicted output is non negetive and continous.

* ReLU is vastly used in NLP.

### Why ReLU widely use
___

* Unlike sigmoid and tanh which sometimes produces tiny gradient and leads to vanishing gradient problem, ReLU produces constant 1 for all positive inputs. This keeps gradients flow through the entire network.

* ReLU doesn't require any expensive operation like exponent. It simply returns the input value if it is positive else 0

* Only a portion of neurons get activated during training which makes it training faster.

* It's linear nature helps to converger faster to the gloabal minima.

# **Pooling in CNN**


After feature maps created from an input, generally it gets bigger in size(dimenstion), memory. Operating process can be slower. This problem is solved with the help of `Pooling`. It's a downsampling technique to reduce the dimension(width and height) of feature maps.


formula:

$$ \frac{(n - k)}{s} + 1 $$

Here,
$$ n= input\ size \\ k= filter\ or\ kernel size \\ s= stride $$

In pytorch,

```python
nn.MaxPool2d(2) # maximum pooling
nn.AvgPool2d(2) # average pooling
```

**Types of Pooling**
---

Generally there are 3 types of pooling in CNN.
* Max Pooling: Maximum value of the feature map is selected.
* Min Pooling: Only minimum value of the feature is selected.
* Average Pooling: Take the average value of the feature map.

Chose of pooling operating depends on the data. Though MaxPooling is most vastly use in real life.


**MaxPooling Vs. AveragePooling**
---

<u>**MaxPooling**</u>

* Takes the maximum value of each region.
* Select the strongest features.

For example, the 4 region of the picture bellow:

<img src="https://i.sstatic.net/6WbWu.png" width= 400/>

* 20 is the maximum value in first region.
* 30 is the maximum value in second region.
* 112 is the maximum value in third region.
* 37 is the maximum value in fourth region.

**Advantage**
* Capture the most prominent feature.
* Works well for detecting shape, corner, edges, texture.
* A small veriance like shifted or distorted can ignored. Can recognize the features.
* For an image, it take the brightest pixels.

**Impact**
* Give concentration on dominant features only.
* Network can learn what is important and what should not be ignored.
* Since only maximum value is taken and small values are discarded, some small spatial values are lost.


<u>**AveragePooling**</u>

* Takes the average value i.e mean value of every regions. For example the feature map bellow have 4 regions.

* Average value of first regions:
$$ = (1 + 4 + 2 + 6) ÷ 4 \\= 3.2 $$

* Average value of others are 6.2, 2.5 and 2.7


**Advantage**
* Preservers overall information.
* Create smooth feature map.
* Reduce the noise.

**Impact**

* Keep all information.
* Network can recognize any features regardless their position.
* Use as the bridge between convolutional layer and fully classification layer.
* If apply to an image, it smooth the picture but sharpe features may not be identified.


**Preventing Overfitting**
---
* Since Pooling reduce the feature map size, so fewer neurons are trained. So model have less chance to memorize.
* Keeps only most important signals, discard the less important features.
* Model becomes less sensitive to small changes like shifting or distorting the image.


**Reducing Computational Complexity**
---
* Fewer the size of feature map, fewer the operating. Makes less computation.
* Memory size get reduces.
* Training gets Faster.

Let's say a 128 * 128 by image become 16 * 16. Nearly 98% value get decreases overall.

### Import necessary libraries

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from torchvision.datasets import FashionMNIST
from torch.utils.data import DataLoader, Subset
from torch.nn.modules.pooling import MaxPool2d
````

````python
torch.manual_seed(42)

device= torch.device("cuda" if torch.cuda.is_available() else 'cpu')
print(f"Connceted Runtime: {device}")
````

### Transform

* Ensure every image is same size i.e (128 𝚇 128)
* Convert every image data to tensor
* Normalize the data in zero centric by assign the mean and standerd deviation 0

````python
transform= transforms.Compose(
    [
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean= [0.5],
            std= [0.5]
        )
    ]
)
````

````python
train_dataset= FashionMNIST(root= "./data", train= True, download= True, transform= transform)
test_dataset= FashionMNIST(root= "./data", train= False, download= True, transform= transform)
````

````python
num_classes= len(train_dataset.classes)
# print(len(train_dataset.classes))
print(f"Number of classes: {num_classes}")
print(f"Number of training samples: {len(train_dataset)}")
print(f"Number of testing samples: {len(test_dataset)}")
````

````python
# train_dataset= Subset(train_dataset, list(range(min(100, len(train_dataset)))))
# test_dataset= Subset(test_dataset, list(range(min(100, len(test_dataset)))))
````

### DataLoader



````python
pin= True if device.type == 'cuda' else False

train_loader= DataLoader(train_dataset, batch_size= 32, shuffle= True, pin_memory= pin)
test_loader= DataLoader(test_dataset, batch_size= 32, shuffle= False, pin_memory= pin)
````

### CNN Class

````python
class MyCNN(nn.Module):
  def __init__(self, num_classes):
    super().__init__()
    self.features= nn.Sequential(
        # Block 1
        nn.Conv2d(1, 32, kernel_size=3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(32),

        nn.Conv2d(32, 32, kernel_size=3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(32),

        nn.MaxPool2d(2),

        # Block 2
        nn.Conv2d(32, 64, kernel_size=3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(64),

        nn.Conv2d(64, 64, kernel_size=3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(64),

        nn.MaxPool2d(2),

        # Block 3
        nn.Conv2d(64, 128, kernel_size= 3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(128),

        nn.Conv2d(128, 128, kernel_size= 3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(128),

        nn.MaxPool2d(2),

        # Block 4
        nn.Conv2d(128, 256, kernel_size= 3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(256),

        nn.Conv2d(256, 256, kernel_size= 3, padding= 'same'),
        nn.ReLU(),
        nn.BatchNorm2d(256),

        nn.MaxPool2d(2)

    )

    self.classifier= nn.Sequential(
        nn.Flatten(),
        nn.Linear(256 * 8 * 8, 256),
        nn.ReLU(),
        nn.Dropout(0.4),

        nn.Linear(256, 256),
        nn.ReLU(),
        nn.Dropout(0.4),

        nn.Linear(256, num_classes)
    )

  def forward(self, X):
    X= self.features(X)
    X= self.classifier(X)

    return X

````

### Training Loop

````python
model= MyCNN(num_classes).to(device)
````

````python
learning_rate= 0.001
epochs= 10
criterion= nn.CrossEntropyLoss()
optimizer= optim.Adam(model.parameters(), lr= learning_rate)
````

````python
for epoch in range(epochs):
  model.train()
  total_loss= 0

  for batch_features, batch_labels in train_loader:
    batch_features= batch_features.to(device)
    batch_labels= batch_labels.to(device)

    outputs= model(batch_features)
    loss= criterion(outputs, batch_labels)

    optimizer.zero_grad()
    loss.backward()

    optimizer.step()
    total_loss += loss.item()
  avg_loss= total_loss / len(train_loader)
  print(f"Epoch: {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")
````

### Evaluation

````python
def evaluate(loader):
  model.eval()
  total, correct= 0, 0
  with torch.no_grad():
    for batch_features, batch_labels in loader:
      batch_features, batch_labels= batch_features.to(device), batch_labels.to(device)

      outputs= model(batch_features)
      _, predicted= torch.max(outputs, 1)
      total += batch_labels.size(0)
      correct += (predicted == batch_labels).sum().item()
  return correct/ total
````

````python
test_acc= evaluate(test_loader)
print(f"Test Accuracy: {test_acc * 100:.3f}%")
````
